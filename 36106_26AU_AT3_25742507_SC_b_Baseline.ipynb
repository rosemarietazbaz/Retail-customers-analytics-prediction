### A.1 Generate Predictions with Baseline Model

```python
# Baseline anomaly detection using K-Nearest Neighbours
baseline_features = [
    'price_difference_ratio',
    'discounted_amount',
    'sub_total'
]

knn = NearestNeighbors(
    n_neighbors=5
)

knn.fit(X_train[baseline_features])

distances, indices = knn.kneighbors(
    X_train[baseline_features]
)

avg_distance = distances.mean(axis=1)

threshold = np.percentile(
    avg_distance,
    95
)

X_train['baseline_anomaly'] = (
    avg_distance > threshold
).astype(int)

X_train['baseline_anomaly'].value_counts()
```

### A.2 Selection of Performance Metrics

> Provide some explanations on why you believe the performance metrics you chose is appropriate


```python
# Calculate baseline anomaly detection metrics
num_anomalies = X_train['baseline_anomaly'].sum()

total_observations = len(X_train)

anomaly_percentage = (
    num_anomalies / total_observations
) * 100

print("Number of anomalies detected:", num_anomalies)

print("Total observations:", total_observations)

print(f"Anomaly percentage: {anomaly_percentage:.2f}%")
```

```python
performance_metrics_explanations = "Traditional supervised learning evaluation metrics such as accuracy, precision, recall, and F1-score were not considered appropriate for this baseline model because anomaly detection is an unsupervised learning task and no labelled anomaly observations were available within the dataset. Instead, the evaluation focused on anomaly prevalence metrics, including the total number of detected anomalies and the percentage of observations classified as anomalous relative to the overall dataset size. The K-Nearest Neighbours baseline model identified 2,705 anomalous observations out of 54,113 total transactions, representing approximately 5.00% of the training dataset. This anomaly proportion was considered reasonable for a baseline anomaly detection approach because anomaly detection models are generally expected to identify a relatively small subset of unusual observations rather than produce balanced class predictions. The selected threshold was based on the 95th percentile of average neighbour distances, allowing the model to identify observations that differ substantially from surrounding transactional patterns while maintaining a manageable anomaly rate. These metrics were considered appropriate because they provide a useful indication of model sensitivity and anomaly prevalence without requiring labelled anomaly data. Monitoring the proportion of detected anomalies also helps assess whether the baseline model behaves reasonably or becomes excessively sensitive to natural transactional variability. One limitation of these evaluation metrics is that they do not directly measure predictive correctness because no ground truth anomaly labels are available for validation. However, they provide an appropriate baseline evaluation framework prior to applying more advanced anomaly detection algorithms."
```
