### B.2 Explore Feature of Interest `name`

```python
print(
    "Unique category names:",
    df['name'].nunique()
)
```

```python
category_name_counts = (
    df['name']
    .value_counts()
)

print(category_name_counts)
```

```python
plt.figure(figsize=(8,4))

sns.countplot(
    x=df['name'],
    order=df['name'].value_counts().index
)

plt.title('Distribution of Product Categories')
plt.xlabel('Category Name')
plt.ylabel('Count')

plt.show()
```

```python
feature_1_insights = "The name feature within the product_category dataset contains four unique product categories: Bikes, Components, Clothing, and Accessories. Each category appears exactly once, confirming that the dataset functions as a clean and well-structured top-level reference table within the retailer’s product classification hierarchy. From a business perspective, these categories represent distinct segments of the retailer’s operations and are likely associated with substantially different pricing structures, inventory requirements, customer purchasing behaviours, and transactional patterns. For example, bicycle products may naturally involve higher-value transactions compared to accessories or clothing items. From a data quality perspective, the absence of duplicated or missing category names indicates strong consistency and reliability within the classification structure. This supports accurate data integration and improves the reliability of downstream preprocessing and modelling tasks. Within the context of anomaly detection, product category information is expected to provide important contextual business segmentation. Transactions considered unusual in one category may represent entirely normal behaviour in another. As a result, category-level context may improve the interpretation of abnormal sales patterns and support more accurate identification of potentially suspicious or operationally inconsistent transactions. One limitation of this feature is that the dataset itself does not contain transactional or quantitative information. Therefore, its analytical value primarily emerges when integrated with product-level and sales transaction datasets during later stages of machine learning modelling."
```
